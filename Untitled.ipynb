```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# --- 1. Load and Prepare the Data ---
print("Loading the merged feature dataset...")
try:
    df = pd.read_csv("data/processed/merged_features_dataset.csv", low_memory=False)
except FileNotFoundError:
    print("Error: 'merged_features_dataset.csv' not found. Please ensure the feature merging script has been run.")
    # Stop execution if file not found
    exit()
    
# Clean data: handle potential errors and missing values
df = df.replace([np.inf, -np.inf], np.nan).dropna()
print(f"Dataset loaded successfully with {len(df)} rows.")

# Separate features (X) and the target label (y)
X = df.drop(columns=["label", "url"], errors="ignore")
y = df["label"]

# --- 2. Encode the Categorical 'tld' Feature ---
# We must convert the text-based TLD into numbers for the model
if 'tld' in X.columns:
    print("Encoding the 'tld' feature...")
    le = LabelEncoder()
    X['tld'] = X['tld'].astype(str) # Ensure all tlds are strings
    X['tld'] = le.fit_transform(X['tld'])
else:
    print("Warning: 'tld' column not found.")

# --- 3. Train a RandomForest Model to Find Important Features ---
print("Training a RandomForest model to determine feature importance...")
# We use a relatively simple model here; its only job is to rank the features.
model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X, y)
print("Model training complete.")

# --- 4. Extract, Rank, and Display the Most Important Features ---
print("\nCalculating feature importances...")
importances = model.feature_importances_
feature_names = X.columns

# Create a DataFrame for better visualization
feature_importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
}).sort_values(by='importance', ascending=False)

# Define how many top features you want to select
N_TOP_FEATURES = 30

# Get the list of the top N features
top_features_list = feature_importance_df.head(N_TOP_FEATURES)['feature'].tolist()

print(f"\n--- Top {N_TOP_FEATURES} Most Important Features ---")
print(top_features_list)

# --- 5. Visualize the Results ---
print("\nGenerating feature importance plot...")
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x='importance', y='feature', data=feature_importance_df.head(N_TOP_FEATURES), palette='viridis', ax=ax)
ax.set_title(f'Top {N_TOP_FEATURES} Feature Importances', fontsize=16)
ax.set_xlabel('Importance Score', fontsize=12)
ax.set_ylabel('Feature Name', fontsize=12)
plt.tight_layout()
plt.show()
```
